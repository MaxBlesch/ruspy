In this notebook I create plots for the simulation results from the ruspy implementation of NFXP and MPEC as well as the solution found by Iskhakov et al. with their NFXP. 
Looking at the variables itself we can see that the three implementations give out almost exactly the same results.
The plots show that the estimates for the maintenance cost parameter are always quite close to the true one while the estimates for the replacement cost vary quite a bit depending on the data set at hand.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.io
```

```python
solution_NFXP_iskhakov = scipy.io.loadmat("solution_NFXP_iskhakov")["result_jr87_97500"]
solution_MPEC = np.load("solution_MPEC.npy")
solution_NFXP = np.load("solution_NFXP.npy")
```

```python
dict = {"MPEC": solution_MPEC, "NFXP": solution_NFXP, "NFXP Iskhakov": solution_NFXP_iskhakov}
for key, value in dict.items():
    plt.plot(value, "r.")
    plt.title(key)
    plt.xlabel("Run")
    plt.ylabel("Cost parameters")
    plt.hlines(11.7257, 0, 250, colors="black")
    plt.hlines(2.4569, 0, 250, colors="black")
    plt.show()
```
